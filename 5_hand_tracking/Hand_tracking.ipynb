```python
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('/home/aiwin/Deep_Learning/hand_tracking/hand_1.jpeg')
```

```python
img
```

```python
plt.imshow(img)
```

```python
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img)
```

```python
mp_hands=mp.solutions.hands
```

```python
hands=mp_hands.Hands(static_image_mode=True,max_num_hands=2,min_detection_confidence=0.3)
# min_detection_confidence=0.3  : accuracy of points
```

```python
mp_drawing=mp.solutions.drawing_utils
```

```python
results=hands.process(rgb_img)
```

```python
print(results.multi_handedness)
```

```python
if results.multi_hand_landmarks:
    for hand_no,hand_landmarks in enumerate(results.multi_hand_landmarks):

        mp_drawing.draw_landmarks(image=rgb_img,landmark_list=hand_landmarks,connections=mp_hands.HAND_CONNECTIONS)

    plt.imshow(rgb_img)
```
